```python
from sklearn.linear_model import LinearRegression
```

```python
#generating data in this code
x = np.array([i for i in range(100)])
y = np.array([i*i for i in range(100)])
```

```python
#splitting the data in this code manually
split = int(0.8*len(x))
x_train,x_test = x[:split],x[split:]
y_train,y_test = y[:split],y[split:]
#reshaping them for 2d input
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

```python
#defining the model
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
# Evaluate the model (optional)
print(f"Model Coefficients: {model.coef_}")
print(f"Model Intercept: {model.intercept_}")
```

```python
# Visualize the results
plt.plot(x, y, color='blue', label='Original Data')  # Original data
plt.plot(x_test, y_pred, color='red', label='Predictions')  # Model predictions
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression on Parabolic Data')
plt.show()
```

```python
#for polnomial expressions
from sklearn.preprocessing import PolynomialFeatures
```

```python
#generating data in this code
x = np.array([i for i in range(100)])
y = np.array([i*i for i in range(100)])
```

```python
#splitting the data in this code manually
split = int(0.8*len(x))
x_train,x_test = x[:split],x[split:]
y_train,y_test = y[:split],y[split:]
#reshaping them for 2d input
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

```python
#Transforming the features using polynomial
degree = 2
poly = PolynomialFeatures(degree)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
#defining the model
model = LinearRegression()
model.fit(x_train_poly,y_train)
y_pred = model.predict(x_test_poly)
```

```python
# Optionally, print model coefficients and intercept
print(f"Model Coefficients: {model.coef_}")
print(f"Model Intercept: {model.intercept_}")
```

```python
# Step 4: Visualize the original data and the model's predictions
plt.plot(x, y, color='blue', label='Original Data')  # Original data points
plt.plot(x_test, y_pred, color='red', label='Polynomial Regression Fit')  # Model predictions
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Polynomial Regression on Parabolic Data')
plt.show()
```

```python
#user input predictions
x_user = np.array([[2]]) #2d array
x_user_poly = poly.transform(x_user)
y_pred_user = model.predict(x_user_poly)
print(f"Predicted value for x = {x_user[0][0]}: {y_pred_user[0]}")
```
